### Analysis
* As expected, the weather becomes significantly warmer as one approaches the equator (0 Deg. Latitude). More interestingly, however, is the fact that the southern hemisphere tends to be warmer this time of year than the northern hemisphere. This may be due to the tilt of the earth.
* There is no strong relationship between latitude and cloudiness. However, it is interesting to see that a strong band of cities sits at 0, 80, and 100% cloudiness.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
api_key = api_keys.api_key
print(api_key)

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
```

```python
# store city information
info_city = []

print ()
print ("-----------------Information is being downloaded---------------------")
print ()

set_info = 1
city_info_record = 1

#Looping all the cities in the list
for i, city in enumerate (cities):
    
    #to have sets of 50 cities each
    if (i % 50 == 0 and i>=50):
        city_info_record = 1
        set_info += 1
        
    #Url to retrive each city info
    url_each_city = url + "&q=" + city
    print ("Processing Record %s of Set %s | %s" % (city_info_record, set_info, city))
    city_info_record += 1
    
    #request API for every list in the list
    try:
        weather_each_city = requests.get(url_each_city).json()
        
        latt_info = weather_each_city["coord"]["lat"]
        long_info = weather_each_city["coord"]["lon"]
        clouds_info = weather_each_city["clouds"]["all"]
        wind_info = weather_each_city["wind"]["speed"]
        temp_max_info = weather_each_city["main"]["temp_max"]
        humidity_info = weather_each_city["main"]["humidity"]
        date_info = weather_each_city["dt"]
        country_info = weather_each_city["sys"]["country"]
        
        
        #add all the information retrives into the list created above
        
        info_city.append({"City": city, 
                          "Latitude": lat_info, 
                          "Longitude": long_info, 
                          "Cloudiness": clouds_info,
                          "Wind Speed": wind_info,
                          "Maximum Temp": temp_max_info,
                          "Humidity": humidity_info,
                          "Date": date_info,
                          "Country": country_info})
    except:
        print ("City not found. Skipping..")
        pass

print ()
print ("---------City weather data download complete-------")
print ()                        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_weatherinfo_pd = pd.DataFrame(info_city)
#Export the city data into a .csv.
city_weatherinfo_pd.to_csv(output_data_file)
#Display the DataFrame
city_weatherinfo_pd.head()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
latt_city = city_weatherinfo_pd["Lattitude"]
Temp_max_city = city_weatherinfo_pd["Maximum Temp"]

plt.scatter(latt_city, Temp_max_city, marker="o", edgecolor="black", linewidths=1, 
            alpha=0.6, label="Cities")

#plot properties
plt.grid(True)
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude (deg)")
plt.title("City Latitude vs. Max Temperature (%s)" % time.strftime("%x"))

plt.savefig("output_data/Figure_Lat_MaxTemp.png")

plt.show()
```
